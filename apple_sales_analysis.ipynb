```python
data.groupby('Product Name').agg({'Star Rating':sum}).sort_values(by='Star Rating',ascending=False)[:10]
```

```python
d1=data.groupby('Product Name')['Star Rating'].sum().sort_values(ascending=False)[:30]
d1
```

THIS CODE REPRESENT THE TOP 10 HIGHEST RATED PRODUCTS

```python
d1.plot(kind='barh',xlabel='product name',ylabel='ratings',title='product rating')
plt.show()
```

```python
data.groupby('Product Name').agg({'Mrp':sum}).sort_values(by='Mrp',ascending=False)[:10]
```

```python
d2=data.groupby('Product Name')['Mrp'].sum().sort_values(ascending=False)[:20]
d2
```

```python
d2.plot(kind='barh',xlabel='price',ylabel='product name',title='top selling phones')
plt.show()
```

THIS GRAPH REPRESENT THE TOTAL SALE OF A TOP 20 PRODUCT 


THE TOP 3 SELLING PHONES ARE 


1.APPLE IPHONE 12 PRO (SIVER , PACIFIC BLUE)


2.APPLE IPHONE 11 PRO(MIDNIGHT GREEN , SPACE GRAY)


3.APPLE IPHONE 12 PRO MAX

```python
data.groupby('Product Name').agg({'Number Of Ratings':sum}).sort_values(by='Number Of Ratings',ascending=False)[:10]
```

```python
d3=data.groupby('Product Name')['Number Of Ratings'].sum().sort_values(ascending=False)[:20]
d3
```
